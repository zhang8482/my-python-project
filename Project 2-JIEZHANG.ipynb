## 7. Customer Segmentation using Clustering:

7.1. Create a matrix RFM (Recency, Frequency, Monetary) for each customer:
- Recency: Number of days since the last purchase
- Frequency: Number of purchases
- Monetary: Total money spent

```python
today = fd_sales['InvoiceDate'].max()
sales_rfm = fd_sales.groupby('CustomerID').agg(
    Recency=('InvoiceDate', lambda date: (today - date.max()).days),
    Frequency=('InvoiceNo', 'nunique'),
    Monetary=('TotalPrice', 'sum')
).reset_index()
```

7.2. Normalize the RFM matrix with StandardScaler:

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
rfm_scaled = std.fit_transform(sales_rfm)
```

7.3. Use KMeans clustering to segment customers into different groups. Determine the optimal number of clusters using the Elbow method.

```python


from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


K = range(1, 11)
inertia = [KMeans(n_clusters=k, random_state=42).fit(rfm_scaled).inertia_ for k in K]


plt.figure(figsize=(8,5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (SSE)')
plt.title('Elbow Method to Determine Optimal k')
plt.show()
```

## 9. Association Rule Mining:
- 9.1 Identify frequently bought products together. Use the Apriori algorithm to extract meaningful association rules.
- 9.2 Based on the rules, suggest product bundling strategies to the retail store.

```python
!pip install mlxtend
!upgrade pip
```

```python

from mlxtend.frequent_patterns import apriori, association_rules

basket = (fd_sales.groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

basket = basket.applymap(lambda x: 1 if x > 0 else 0)

frequent_items = apriori(basket, min_support=0.02, use_colnames=True)

rules = association_rules(frequent_items, metric="lift", min_threshold=1)

print(rules.sort_values('lift', ascending=False).head(10))
```

```python
import networkx as nx

rules_top = rules.sort_values('lift', ascending=False).head(15)

G = nx.DiGraph()

for _, row in rules_top.iterrows():
    for antecedent in row['antecedents']:
        for consequent in row['consequents']:
            G.add_node(antecedent, color='lightblue')
            G.add_node(consequent, color='lightgreen')
            G.add_edge(antecedent, consequent, weight=row['lift'])

edges = G.edges()
weights = [G[u][v]['weight'] for u, v in edges]

plt.figure(figsize=(12,8))
pos = nx.spring_layout(G, k=0.5, seed=42)
nx.draw_networkx_nodes(G, pos, node_size=1800, node_color='skyblue')
nx.draw_networkx_edges(G, pos, edgelist=edges, width=[w for w in weights], alpha=0.6, edge_color='gray', arrows=True)
nx.draw_networkx_labels(G, pos, font_size=8)
plt.title("Association Rules Network (Top 15 by Lift)", fontsize=14)
plt.axis('off')
plt.show()
```

## 14. Product Recommendation Systems:
- Develop a system to recommend products to users.
- Consider collaborative filtering techniques, matrix factorization, or deep learning-based approaches like neural collaborative filtering.

```python
from sklearn.metrics.pairwise import cosine_similarity

user_product_matrix = fd_sales.pivot_table(index='CustomerID', columns='StockCode', values='Quantity', fill_value=0)
cosine_sim = cosine_similarity(user_product_matrix)

def get_recommendations(user_id, cosine_sim=cosine_sim):
    idx = user_product_matrix.index.get_loc(user_id)
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    user_idx = sim_scores[1][0]
    similar_user_products = user_product_matrix.iloc[user_idx]
    recommended_products = similar_user_products[similar_user_products > 0].index.tolist()
    return recommended_products
```

## 15. Churn Prediction:
- Predict if a customer will stop buying products in the near future.
- Features can include Recency, Frequency, Monetary values, average time between purchases, total categories bought, etc.

```python
from sklearn.linear_model import LogisticRegression

fd_sales['LastPurchase'] = fd_sales.groupby('CustomerID')['InvoiceDate'].transform('max')
max_date = fd_sales['InvoiceDate'].max()
fd_sales['Churn'] = (max_date - fd_sales['LastPurchase']).dt.days > 180

X = fd_sales[['TotalPrice', 'Quantity', 'ReturnFlag']]
y = fd_sales['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
clf = LogisticRegression()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
